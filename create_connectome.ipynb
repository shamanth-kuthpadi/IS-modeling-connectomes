```python
import pandas as pd
import numpy as np

# Load the CSV file into a DataFrame
file_path = 'data/connectome_01.csv'
connectome_df = pd.read_csv(file_path, header=None)
print(connectome_df.head())
# Convert DataFrame to a NumPy array for easier manipulation
connectome_matrix = connectome_df.to_numpy()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(connectome_matrix, cmap="coolwarm", square=True)
plt.title('Structural Connectome Heatmap')
plt.xlabel('Brain Region')
plt.ylabel('Brain Region')
plt.show()
```

```python
import networkx as nx

# Create a graph from the adjacency matrix
G = nx.from_numpy_array(connectome_matrix)

# Plot the graph
plt.figure(figsize=(10, 10))
pos = nx.spring_layout(G)  # Or try a circular or shell layout for brain-like patterns
nx.draw(G, pos, with_labels=False, node_size=100, node_color='skyblue', edge_color='gray')
plt.title('2D Structural Connectome Graph')
plt.show()
```
