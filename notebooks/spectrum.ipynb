```python
connectome = Connectome(file='/Users/shamanthk/Documents/Research/IS-modeling-connectomes/repeated_10_scale_500/103212_repeated10_scale500.graphml', label='103212')
```

```python
net = connectome.read_net(use_3d=True)
evals, evecs, map = spectrum(net)
```

## Adjacency Matrix Spectrum

```python
import seaborn as sns

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(evals, marker='o', color='b', linestyle='-')
plt.title("Eigenvalues in Decreasing Order")
plt.xlabel("Index")
plt.ylabel("Eigenvalue")
plt.grid(True)

plt.subplot(1, 2, 2)
sns.heatmap(evecs, cmap='plasma', annot=False, cbar=True)
plt.title("Eigenvector Spectrum")
plt.xlabel("Eigenvector Index")
plt.ylabel("Vector Components")

plt.tight_layout()
plt.show()
```

## Laplacian Matrix Spectrum

```python
evals, evecs, map = spectrum(net, lp=True)
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(evals, marker='o', color='b', linestyle='-')
plt.title("Eigenvalues in Decreasing Order")
plt.xlabel("Index")
plt.ylabel("Eigenvalue")
plt.grid(True)

plt.subplot(1, 2, 2)
sns.heatmap(evecs, cmap='plasma', annot=False, cbar=True)
plt.title("Eigenvector Spectrum")
plt.xlabel("Eigenvector Index")
plt.ylabel("Vector Components")

plt.tight_layout()
plt.show()
```
